```python
import matplotlib.dates as dates
import matplotlib.pyplot as plt
```

```python
idx = mcdon.loc['2007-01-01': '2007-05-01'].index
stock = mcdon.loc['2007-01-01': '2007-05-01', 'Adj. Close']

fig, ax = plt.subplots(figsize=(15,10))
ax.plot_date(idx, stock, '-')

ax.yaxis.grid(True) # for gridlines but ggplot does this for us
ax.xaxis.grid(True)
fig.autofmt_xdate()
```

```python
stock = mcdon.loc['2007-01-01': '2007-05-01', 'Adj. Close']

fig, ax = plt.subplots(figsize=(15,10))
ax.plot_date(idx, stock, '-')

# LOCATING
ax.xaxis.set_major_locator(dates.MonthLocator())

# FORMATTING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n%Y-%m'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0)) # setting it to certain weekday 0=Monday
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
fig.autofmt_xdate()
```
