```python
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(1,5,11)
y = x ** 2
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('TITLE')
plt.show()
```

```python
row_num = 1
col_num = 2
plot_number = 1
plt.subplot(row_num, col_num, plot_number)
plt.plot(x,y,'r')

plot_number = 2
plt.subplot(row_num, col_num, plot_number)
plt.plot(x,y,'b')
plt.show()
```

```python
# OBJECT ORIENTED VERSION
fig = plt.figure() # creates the canvas
left = 0.1
bottom = 0.1
width = 1.0
height = 1.0

axes_arg_list = [left, bottom, width, height] # each arg is the percent of the canvas to take up
axes = fig.add_axes(axes_arg_list)
axes.plot(x,y)
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('TITLE')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.3, 0.4, 0.4, 0.3])

axes1.plot(x,y, 'r')
axes1.set_title('LARGE PLOT')
axes2.plot(y,x, 'b')
axes2.set_title('small plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'r')
axes[0].set_title('first plot')
axes[1].plot(y,x,'b')
axes[1].set_title('second plot')

plt.tight_layout()
```

# Figure Size and DPI (dots per inch or pixels per inch)

```python
fig = plt.figure(figsize=(8,8), dpi=500)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8,8), dpi=100)

axes[1,1].plot(x,y)
plt.tight_layout()
```

```python
# Legends
fig = plt.figure(figsize=(8,8), dpi=100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, np.log(x), 'b', label='log x')
ax.plot(x, x**2, label='x-squared')

ax.legend(loc='best') # this will check for labels on each plotted line
```

```python
fig = plt.figure(figsize=(6,6), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,np.log(x), color='red', linewidth=10, alpha=0.2) # linewidth is a multiple of the default thickness
# alpha is the transparency
```

```python
fig = plt.figure(figsize=(6,6), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,np.log(x), color='red', lw=10, linestyle='dotted', alpha=0.2) # lw is shorthand for linewidth
# alpha is the transparency
```

```python
fig = plt.figure(figsize=(6,6), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,np.log(x), color='red', lw=3, ls='--', alpha=0.2, marker='o', markersize=10,
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='g') # ls is shorthand for linestyle
# alpha is the transparency
```

```python
fig = plt.figure(figsize=(4,4), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,np.log(x), color='red', lw=3, ls='--', alpha=0.2, marker='o', markersize=10,
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='g') # ls is shorthand for linestyle

ax.set_ylim([0,3])
ax.set_xlim([0,10])
```
